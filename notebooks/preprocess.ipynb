```python
import os
import pandas as pd
import json
```

### Valentine

```python
base_path = '../datasets/ChEMBL'
use_cases = ['Joinable', 'Semantically-Joinable', 'Unionable', 'View-Unionable']
```

```python
unique_columns = {}

for use_case in use_cases:
    folder_path = os.path.join(base_path, use_case, 'target-tables')

    if os.path.exists(folder_path):
        for file_name in os.listdir(folder_path):
            if file_name.endswith('.csv'):
                file_path = os.path.join(folder_path, file_name)
                try:
                    df = pd.read_csv(file_path)
                    for column in df.columns:
                        if column not in unique_columns:
                            value_counts = df[column].value_counts()
                            if len(value_counts) > 100:
                                value_counts = value_counts.head(50)
                            unique_columns[column] = value_counts.index.tolist()
                except Exception as e:
                    print(f"Error reading {file_path}: {e}")
    else:
        print(f"Folder not found: {folder_path}")
unique_columns
```

```python
file_name = 'chembl_unique_columns.json'
with open(file_name, 'w') as f:
    json.dump(unique_columns, f, indent=2)
    print(f"Unique columns saved to {file_name}")
```

### GDC
